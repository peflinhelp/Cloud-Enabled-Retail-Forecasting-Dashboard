```python

prefix_unique_count = len(merged_df_cleaned['prefix'].unique())
item_name_unique_count = len(merged_df_cleaned['item_name'].unique())
print (f"{prefix_unique_count} and {item_name_unique_count}") # we can see that we succesfully created the prefix column
```

```python
mapping =(
    merged_df_cleaned.dropna(subset=['item_name'])
      .groupby('prefix')['item_name']
      .agg(lambda x: x.mode()[0])
      .to_dict()
)

mapping
```

```python
merged_df_cleaned['item_name'] = merged_df_cleaned.apply(
    lambda row: mapping.get(row['prefix'], row['item_name'])
                if pd.isna(row['item_name']) or row['item_name'] == ''
                else row['item_name'],
    axis=1
)

merged_df_cleaned[merged_df_cleaned['ITEMID']=='tro23wfga23562tm1'] # all missing values has been imputed
```

```python
merged_df_cleaned = merged_df_cleaned.drop(columns=['prefix'])
```

# daily revenue predictions


```python
df_daily_revenue = (
    sales_df.groupby('Date Timesold')
            .agg(total_revenue=('NETAMOUNTINCLTAX', 'sum'))
            .reset_index()
            .rename(columns={'Date Timesold': 'date'})
)
# summing netamountincltax same as sum*quatity sold
sales_df.info()
```

```python
df_daily_revenue['date'] = pd.to_datetime(df_daily_revenue['date'], errors='coerce')
df_daily_revenue = df_daily_revenue.sort_values('date').set_index('date')

from google.colab import drive
drive.mount('/content/drive')

merged_df_cleaned.to_excel('/content/drive/MyDrive/final_table.xlsx', index=False)
```

```python
df_daily_revenue.rename(columns={'total_revenue': 'daily_total_rev'}, inplace=True)  # we need date as an index for proper shifting
df_daily_revenue
```

```python
#we're goinf to train three models random forest, sarima, and prophet
# feature engeenering, creating lag features and MA(moving average) features, this is only for random forest, sarima and prophet allready can handle time series properly.
df_daily_revenue['lag1'] = df_daily_revenue['daily_total_rev'].shift(1)
df_daily_revenue['lag7'] = df_daily_revenue['daily_total_rev'].shift(7)
df_daily_revenue['lag30'] = df_daily_revenue['daily_total_rev'].shift(30)

df_daily_revenue['roll_mean_7']  = df_daily_revenue['daily_total_rev'].rolling(7).mean()
df_daily_revenue['roll_mean_30'] = df_daily_revenue['daily_total_rev'].rolling(30).mean()

# Drop rows with NaN after shifts
df_feat = df_daily_revenue.dropna()
```

```python
df_daily_revenue
# as we see 4 columns were added. lag allows to caputer temporal dependency (autocorrelation). MA smooth the series to reduce noise and reveal trends (simply this mean of 30 days)
# we have 365 day, and all lag and ma values are placed correctly.
```

```python
# train and test spliting is the same for all daily revenue predictive models
index_80_percent = int(len(df_daily_revenue)*0.8) # t his gives an int which is 80 % of our data, basicaly we're going to use index_80_percent to split data into training set and testing set
test_set_rf = df_daily_revenue.iloc[index_80_percent:] # 20% test | RF
train_set_rd = df_daily_revenue.iloc[:index_80_percent] # 80% train | RF


train_set_sarima = df_daily_revenue.iloc[:index_80_percent] # 80% test | sarima
test_set_sarima  = df_daily_revenue.iloc[index_80_percent:] # 20% train | sarima
```

# Weekly revenue prediction | training  after Daily revenue models


```python

sales_df['year_week'] = sales_df['Date Timesold'].dt.to_period('W').astype(str)

df_weekly_revenue = (
    sales_df.groupby('year_week')
            .agg(total_revenue=('NETAMOUNTINCLTAX', 'sum'))
            .reset_index()
)
df_weekly_revenue.rename(columns={'total_revenue': 'weekly_total_rev'}, inplace=True)
df_weekly_revenue.head()
```
